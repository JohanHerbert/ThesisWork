```python
import yfinance as yf
import datetime
import pandas as pd
```

```python

# # Set the date range
# end_date = datetime.datetime.today()
# start_date =  #end_date - datetime.timedelta(days=30*365)

end_date = datetime.datetime.today()
start_date = datetime.datetime(1997, 5, 15)


# Download 20 years of data for AAPL and AMZN
tickers = ['AAPL', 'AMZN']
data = yf.download(tickers, start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))

# Save to CSV
reducedData = data.iloc[:,0:2]
reducedData.columns = ['AAPL', 'AMZN']
reducedData.to_csv('apple_amazon_since_1997.csv', index=True)
```

```python
readData = pd.read_csv('apple_amazon_since_1997.csv', index_col=0)
readData.plot(ylabel='Price',  figsize=(15, 5)) 
```
